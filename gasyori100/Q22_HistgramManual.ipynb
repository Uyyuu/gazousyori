Q22 ヒストグラム操作

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori_dark = cv2.imread("imori_dark.jpg").astype(float)

def show_image(a):
#     img_rgb = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
# よくわからんエラー出たから自分でRGBに入れ替えた
    img_rgb = a[:,:,(2,1,0)]
    plt.imshow(img_rgb)
    plt.show()

def histgram(h_img):
    plt.hist(h_img.ravel(), bins=255, range=(0, 255), rwidth=0.7 )
    plt.show()
```

```python
def hist_manual(image,s0,m0):
    img = image.copy()
    s = img.std(ddof=0)
    m = img.mean()
    img = (s0/s)*(img-m)+m0
    img = np.clip(img.astype(int),0,255)
    
    return img
```

```python
q22_img = hist_manual(imori_dark,52,128)
show_image(q22_img)
```

```python
histgram(q22_img)
```
