```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori = cv2.imread("imori.jpg")

def show_image(a):
    img_rgb = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.show()
```

```python
def conditions(RGB):
    if 0<= RGB <64:
        return 32
    elif 64<= RGB < 128:
        return 96
    elif 128<= RGB < 192:
        return 160
    else:
        return 224

def quantization(img):
    q_img = img.copy()
    b = img[:,:,0].copy()
    g = img[:,:,1].copy()
    r = img[:,:,2].copy()
    h,w,_ = img.shape
    
    q_b = list(map(conditions, np.array(b).reshape(-1,1)))
    q_g = list(map(conditions, np.array(g).reshape(-1,1)))
    q_r = list(map(conditions, np.array(r).reshape(-1,1)))
    
    q_img[:,:,0] = list(np.array(q_b).reshape(h,w))
    q_img[:,:,1] = list(np.array(q_g).reshape(h,w))
    q_img[:,:,2] = list(np.array(q_r).reshape(h,w))
    
    return q_img

def answer(img):
    out = img.copy()
    out = out // 64 * 64 + 32
    return out
```

```python
quan_image = quantization(imori)
show_image(quan_image)
```

```python
answer_image = answer(imori)
show_image(answer_image)
```
