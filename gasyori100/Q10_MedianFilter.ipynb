Q10 メディアンフィルタ

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori = cv2.imread("imori.jpg")
noise_imori = cv2.imread("imori_noise.jpg")

def show_image(a):
    img_rgb = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.show()
```

```python
def padding_0(src):
    return cv2.copyMakeBorder(src,1,1,1,1,cv2.BORDER_CONSTANT,(0,0,0))

def median_filter(img):
    image = img.copy()
    pad_image = padding_0(img)
    h,w,_ = img.shape
    
    for i in range(1,w):
        for j in range(1,h):
            b = pad_image[i-1:i+2,j-1:j+2,0]
            g = pad_image[i-1:i+2,j-1:j+2,1]
            r = pad_image[i-1:i+2,j-1:j+2,2]
            
            image[i-1,j-1,0] = np.median(b)
            image[i-1,j-1,1] = np.median(g)
            image[i-1,j-1,2] = np.median(r)
    return image
```

```python
q10_img = median_filter(noise_imori)
show_image(q10_img)
```
