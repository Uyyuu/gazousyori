Q28 アフィン変換（平行移動）

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori = cv2.imread("imori.jpg").astype(float)

def show_image(a):
    img_rgb = a[:,:,(2,1,0)]
    plt.imshow(img_rgb)
    plt.show()

def histgram(h_img):
    plt.hist(h_img.ravel(), bins=255, range=(0, 255), rwidth=0.7 )
    plt.show()
```

```python
def AffineTranslate(image,tx,ty):
    img = image.copy()
    h,w,_ = img.shape
    out = np.zeros((h,w,3))
    original = np.array([0,0])
    
    for i in range(h):
        for j in range(w):
            for k in range(3):
                original = np.array([j,i]) - np.array([tx,ty])
                if 0 <= original[0] < h and 0 <= original[1] < w:
                    out[i,j,k] = img[original[1],original[0],k]

    return out.astype(int)
```

```python
q28_img = AffineTranslate(imori,30,-30)
show_image(q28_img)
```
