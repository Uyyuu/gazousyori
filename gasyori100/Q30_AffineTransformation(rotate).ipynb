Q30 アフィン変換（回転）

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori = cv2.imread("imori.jpg").astype(float)

def show_image(a):
    img_rgb = a[:,:,(2,1,0)]
    plt.imshow(img_rgb)
    plt.show()

def histgram(h_img):
    plt.hist(h_img.ravel(), bins=255, range=(0, 255), rwidth=0.7 )
    plt.show()
```

```python
def AffineRotate(image,tx,ty,theta):
    img = image.copy()
    h,w,_ = img.shape
    H = int(h)
    W = int(w)
    out = np.zeros((H,W,3))
    original = np.array([0,0])
    t = -np.pi*theta/180
    
    for i in range(H):
        for j in range(W):
            for k in range(3):
                original = np.array([[np.cos(t),np.sin(t)],[-np.sin(t),np.cos(t)]])\
                                        @np.array([j,i]) - np.array([tx,ty])
                if 0 <= original[0] < w and 0 <= original[1] < h:
                    out[i,j,k] = img[int(original[1]),int(original[0]),k]

    return out.astype(int)
```

```python
q30_img = AffineRotate(imori,0,0,30)
show_image(q30_img)
```

```python
q30_img_1 = AffineRotate(imori,-40,30,30)
show_image(q30_img_1)
```
