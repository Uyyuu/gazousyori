```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori_gamma = cv2.imread("imori_gamma.jpg").astype(float)

def show_image(a):
    img_rgb = a[:,:,(2,1,0)]
    plt.imshow(img_rgb)
    plt.show()

def histgram(h_img):
    plt.hist(h_img.ravel(), bins=255, range=(0, 255), rwidth=0.7 )
    plt.show()
```

```python
def gamma(image,g):
    img = image.copy()/255
    img = (img**(1/g))*255
    img = img.astype(int)
    
    return img
```

```python
q24_img = gamma(imori_gamma,2.2)
show_image(q24_img)
```

```python
show_image(imori_gamma.astype(int))
```
