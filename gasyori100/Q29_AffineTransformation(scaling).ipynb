Q29 アフィン変換（拡大・平行移動）

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori = cv2.imread("imori.jpg").astype(float)

def show_image(a):
    img_rgb = a[:,:,(2,1,0)]
    plt.imshow(img_rgb)
    plt.show()

def histgram(h_img):
    plt.hist(h_img.ravel(), bins=255, range=(0, 255), rwidth=0.7 )
    plt.show()
```

```python
def AffineScaling(image,tx,ty,ax,ay):
    img = image.copy()
    h,w,_ = img.shape
    H = int(h*ay)
    W = int(w*ax)
    out = np.zeros((H,W,3))
    original = np.array([0,0])
    
    for i in range(H):
        for j in range(W):
            for k in range(3):
                original = 1/(ax*ay)*np.array([[ay,0],[0,ax]])@np.array([j,i]) - np.array([tx,ty])
                if 0 <= original[0] < w and 0 <= original[1] < h:
                    out[i,j,k] = img[int(original[1]),int(original[0]),k]

    return out.astype(int)
```

```python
q29_img_1 = AffineScaling(imori,0,0,1.3,0.8)
show_image(q29_img_1)
```

```python
q29_img_2 = AffineScaling(imori,30,-30,1.3,0.8)
show_image(q29_img_2)
```
