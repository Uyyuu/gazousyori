Q13 Max-Minフィルタ

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori = cv2.imread("imori.jpg")
noise_imori = cv2.imread("imori_noise.jpg")

def show_image(a):
    img_rgb = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.show()

def show_gray(gray_img):
    plt.imshow(gray_img)
    plt.gray()
    plt.show()
```

```python
def padding_0(src):
    return cv2.copyMakeBorder(src,1,1,1,1,cv2.BORDER_CONSTANT,(0,0,0))

def gray(gray_image):
    b = gray_image[:,:,0].copy()
    g = gray_image[:,:,1].copy()
    r = gray_image[:,:,2].copy()
    
    y = 0.2126*r + 0.7152*g + 0.0722*b
    y = y.astype(int)
    
    return y    

def Max_Min_filter(img):
    image = gray(img)
    pad_image = padding_0(image)
    h,w = image.shape
   
    for i in range(1,w):
        for j in range(1,h):
            pic = pad_image[i-1:i+2,j-1:j+2]
            image[i-1,j-1] = np.max(pic) - np.min(pic)
           
    return image
```

```python
q13_img = Max_Min_filter(imori)
show_gray(q13_img)
```
