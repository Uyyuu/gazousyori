最近傍補間法

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

imori = cv2.imread("imori.jpg").astype(float)

def show_image(a):
    img_rgb = a[:,:,(2,1,0)]
    plt.imshow(img_rgb)
    plt.show()

def histgram(h_img):
    plt.hist(h_img.ravel(), bins=255, range=(0, 255), rwidth=0.7 )
    plt.show()
```

```python
def nearest(image,scale=1.5):
    img = image.copy()
    h,w,_ = img.shape
    H = int(h*scale)
    W = int(w*scale)
    out = np.zeros((H,W,3), dtype=int)
    for i in range(H):
        for j in range(W):
            for k in range(3):
                out[i,j,k] = img[round(i/scale),round(j/scale),k]
    return out
```

```python
q25_img = nearest(imori)
show_image(q25_img)
```
