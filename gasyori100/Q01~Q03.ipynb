```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
img = cv2.imread("imori.jpg")
# BGRの順で保存されてる
# cv2.imshow("imori",img)
```

```python
def show_image(a):
    img_rgb = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.show()
```

```python
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# plt.imshow(img)
# plt.show()
show_image(img)
```

```python
img_t = img.copy()
h,w,c = img_t.shape
img_t[:h//2, :w//2] = img_t[:h//2, :w//2, (2, 1, 0)]
show_image(img_t)
```

```python
# Q1
img1 = img.copy()
img1[:,:] = img1[:,:,(2,1,0)]
show_image(img1)
```

```python
# Q2
def gray(gray_image):
    b = gray_image[:,:,0].copy()
    g = gray_image[:,:,1].copy()
    r = gray_image[:,:,2].copy()
    
    y = 0.2126*r + 0.7152*g + 0.0722*b
    y = y.astype(int)
    
    return y

gray_img = img.astype(float)
gray_img_out = gray(gray_img)
plt.imshow(gray_img_out)
plt.gray()
plt.show()
```

```python
# Q3
def binarized(binarized_image):
    b = binarized_image[:,:,0].copy()
    g = binarized_image[:,:,1].copy()
    r = binarized_image[:,:,2].copy()
    
    y = 0.2126*r + 0.7152*g + 0.0722*b
    
    for i in range(0,binarized_image.shape[0]):
        for j in range(0,binarized_image.shape[1]):
            if y[i][j] < 128:
                y[i][j] = 0
            else:
                y[i][j] = 255
    
    return y
    
bina_img = binarized(img)
plt.imshow(bina_img)
plt.gray()
plt.show()
```
